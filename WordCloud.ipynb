```python
# import package
!pip install --target=$nb_path wikipedia
import wikipedia
# title of the wiki page
covid19 = wikipedia.page('COVID-19')
text = covid19.content
# Import package
import re
```

**Cleaning Data**

```python
# Cleaning Data
text
```

```python
text = text.replace('\n', '')
text = text.replace(',', '')
text = text.replace('.', '')
text
# How do I use re.sub()
```

**Word Cloud**

```python
# Import Package
import matplotlib.pyplot as plt

# Function to plot the word cloud
def plot_cloud(wordcloud):
  plt.figure(figsize = (40,30)) #set fig size
  plt.imshow(wordcloud) #display image
  plt.axis("off"); #leave out axis details
```

**Plotting the Word Cloud**

```python
from wordcloud import WordCloud, STOPWORDS # Importing the package in full
wordcloud = WordCloud(width = 4000, height = 2000, 
                      random_state=1, background_color = 'white',
                      colormap = 'Set2', collocations = False,
                      stopwords = STOPWORDS).generate(text)

plot_cloud(wordcloud)
```
